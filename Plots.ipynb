```python
import numpy as np
```

# 400 Тел

```python
time_400 = []
with open('time400.txt') as f:
    for line in f.readlines():
        time.append(np.mean(list(map(float, line.split()))))
print(time)
```

```python
threads = list(range(1, 11))
```

# Время работы программы от количества потоков

```python
import matplotlib.pyplot as plt

plt.plot(threads, time_400)
plt.xlabel('Количество потоков')
plt.ylabel('Время работы[сек]')
plt.grid()
plt.show()
```

```python
time_seq_400 = np.mean([77.042625, 76.149509, 75.529898])
speedup_400 = [time_seq_400 / t for t in time_400]
```

```python
print(speedup_400)
```

# Ускорение от количества потоков

```python
plt.plot(threads, speedup_400)
plt.xlabel('Количество потоков')
plt.ylabel('Ускорение')
plt.grid()
plt.show()
```

```python
effective_400 = [s / p for s, p in zip(speedup_400, threads)]
print(effective_400)
```

# Эффективность от количества потоков

```python
plt.plot(threads, effective_400)
plt.xlabel('Количество потоков')
plt.ylabel('Эффективность')
plt.grid()
plt.show()
```
